```python
import pandas as pd

# 파일명에 맞게 로드
filename = "naver_place_reviews_20250625_123653.xlsx"
df = pd.read_excel(filename)

df.head()
```

```python
# 이모티콘 정리

import unicodedata
import pandas as pd

def remove_emoji_only(text):
    if pd.isna(text):
        return ""

    return ''.join(
        ch for ch in str(text)
        if not unicodedata.category(ch).startswith('So')  # Symbol, Other
        and not unicodedata.name(ch, "").startswith(("EMOJI", "FACE WITH", "GRINNING", "SMILING"))
    )
```

```python
df["content"] = df["content"].apply(remove_emoji_only)

df
```

```python
# content 빈 공란 제거

import re

def is_meaningful_text(text):
    if not isinstance(text, str):
        return False
    return bool(re.search(r'[가-힣a-zA-Z0-9]', text))

df = df[df["content"].apply(is_meaningful_text)]
```

```python
df
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import torch.nn.functional as F
from tqdm import tqdm

# 모델명 (한국어 감정 분석)
model_name = "nlp04/korean_sentiment_analysis_kcelectra"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# GPU 사용 가능 시 적용
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)
model.eval()

# 감정 분류 함수 정의
def classify_sentiment(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True).to(model.device)
    with torch.no_grad():
        outputs = model(**inputs)
        probs = F.softmax(outputs.logits, dim=-1)
        pred = torch.argmax(probs, dim=-1).item()
    return "긍정" if pred == 1 else "부정"
```

```python
from tqdm import tqdm

# tqdm 사용 시 DataFrame progress 표시
tqdm.pandas()

# 감정 분석 결과 컬럼 추가
df['sentiment'] = df['content'].progress_apply(classify_sentiment)
```

```python
positive_reviews = df[df['sentiment'] == '긍정']
negative_reviews = df[df['sentiment'] == '부정']
```

```python
import pandas as pd
import numpy as np

# ===== 0) 데이터 로드 =====
df = pd.read_excel("Review_good_bad.xlsx")

# ===== 1) sentiment 정규화: 긍정=1, 부정=0 =====
def normalize_sentiment(x):
    if pd.isna(x):
        return np.nan
    if isinstance(x, (int, float)):
        # 1/0/-1 등 숫자형
        return 1 if x == 1 else 0
    s = str(x).strip().lower()
    if s in ["긍정"]:
        return 1
    if s in ["부정"]:
        return 0
    return np.nan

df["sentiment_bin"] = df["sentiment"].apply(normalize_sentiment)
```

```python
df.head()
```

```python
df.fillna(0, inplace=True)
df.head()
```

```python
df.info()
```

```python
# 3) 키워드 목록
keywords = [
    "자연스러움", "화사함", "빈티지", "모노크롬", "선명함", "아날로그", "파스텔",
    "아이돌", "캐릭터", "브랜드", "계절", "미디어",
    "기타", "의상", "헤어", "안경", "식물",
    "여유로움", "조용함", "2인용", "단체", "접근성", "청결", "인기"
]
```

```python
# ===== 3) 비율 -> 1~5점 변환 =====
def ratio_to_score(p):
    if pd.isna(p):
        return np.nan
    if p <= 0.20:
        return 1
    elif p <= 0.40:
        return 2
    elif p <= 0.60:
        return 3
    elif p <= 0.80:
        return 4
    else:
        return 5

# ===== 4) (1) 전체 리뷰 대비 긍정 점수 =====
# 분모: 해당 place의 전체 리뷰 수(행 수). 분자: sentiment_bin==1 개수.
overall_rows = []
for place, g in df.groupby("place_id"):
    total_reviews = len(g)                         # 전체 리뷰 수
    pos_reviews   = (g["sentiment_bin"] == 1).sum()  # 긍정 리뷰 수(결측 제외 자동 처리)
    # 전체 리뷰 = 분모, 결측 sentiment는 자연스럽게 분모엔 포함되지만 분자엔 포함 안 됨
    pos_ratio_all = np.nan if total_reviews == 0 else pos_reviews / total_reviews
    overall_rows.append({
        "place_id": place,
        "긍정비율": pos_ratio_all,
        "사진관점수": ratio_to_score(pos_ratio_all)
    })

overall_scores = pd.DataFrame(overall_rows)

overall_scores.head()
```

```python
# ===== 5) (2) 긍정 리뷰 대비 키워드별 점수 =====
# 분모: 해당 place의 '긍정 리뷰' 수. 분자: (긍정 리뷰 & 해당 키워드=1) 수.
kw_rows = []
for place, g in df.groupby("place_id"):
    # place 내 긍정 리뷰만 필터
    g_pos = g[g["sentiment_bin"] == 1]
    pos_total = len(g_pos)  # 긍정 리뷰 수 (분모)

    row = {"place_id": place}
    for kw in keywords:
        if kw not in g.columns:
            row[kw] = np.nan
            continue

        # 긍정 리뷰 중 키워드가 1인 리뷰 수 (분자)
        pos_kw = (g_pos[kw] == 1).sum()

        if pos_total == 0:
            ratio = np.nan
        else:
            ratio = pos_kw / pos_total  # 긍정 리뷰 대비 언급 비율

        row[kw] = ratio_to_score(ratio)  # 1~5점 변환
    kw_rows.append(row)

keyword_scores = pd.DataFrame(kw_rows)

keyword_scores
```

```python
# place_id 기준 병합
merged_scores = pd.merge(overall_scores, keyword_scores, on="place_id", how="outer")

# 결과 확인
merged_scores.head()
```

```python
# place_id별 대표 place_name 추출
place_info = df.groupby("place_id")["place_name"].first().reset_index()

# merged_scores와 병합
merged_with_name = pd.merge(place_info, merged_scores, on="place_id", how="left")
merged_with_name.head()
```

```python
# # ': 네이버' 문자열 제거
merged_with_name["place_name"] = merged_with_name["place_name"].str.replace(r"\s*:\s*네이버\s*", "", regex=True)

merged_with_name
```

```python
# 6) 결과 저장
merged_with_name.to_excel("place_keyword_scores.xlsx", index=False)
print("저장 완료: place_keyword_scores.xlsx")
```
